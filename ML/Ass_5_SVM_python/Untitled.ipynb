```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/home/test/BE/CL-7_practice/ML/Datasets/Wine Dataset/wine.data",sep=',')
```

```python
x=data.iloc[:,1:14]
y=data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
def create_network():
    model=Sequential()
    model.add(Dense(8,activation='relu',input_dim=13))
    model.add(Dense(4,activation='relu'))
    model.add(Dense(1,activation='relu'))
    
    model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
    return model
```

```python
classifier=create_network()
classifier.fit(x_train,y_train,epochs=500,batch_size=25)
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
score=classifier.evaluate(x_test,y_test)
score
```
